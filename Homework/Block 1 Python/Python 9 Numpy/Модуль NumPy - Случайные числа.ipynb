```python
import numpy as np

np.random.rand() # генерируем случайное число в диапазоне от 0 и до 1.
```

```python
np.random.rand() * 100 # генерируем случайное число в диапазоне от 0 и до 100.
```

```python
np.random.rand(5) # функция rand() принимает размерность итогового массива, когда передают аргумент.
```

```python
np.random.rand(2, 3) # несколько аргументов, создают многомерные массивы.
```

```python
np.random.rand(2, 3, 4, 10, 12, 23) # создаем шестимерный массив
```

```python
shape = (3, 4)
np.random.rand(shape) # картеж не может быть интерпритирован как интеджер int.
```

```python
shape = (3, 4)
np.random.rand(*shape) # распаковываем картеж и передает в виде аргументов, это не совсем корректный вид.
```

```python
shape = (3, 4)
np.random.sample(shape) # но функция Sample как раз и принимает форму массива в виде кортежа.
```

```python
np.random.uniform() # функция юниформ без аргументов генерит случайное число от 0 и до 1.
```

```python
np.random.uniform(-30, 50) # функция юниформ принимает граничные значения, т.е. значения что нужны для генерации числа. 
```

```python
np.random.uniform(0.5, 0.75, size=5) # с помощью аргумента size можно задавать форму массива в диапазоне,  или одно число
```

```python
np.random.uniform(-1000, 500, size=(2, 3)) # с помощью аргумента size можно задавать форму массива в диапазоне, или кортежа
# тут могут быть целые числа и числа с плавующей точкой
```

```python
np.random.randint(4, size=(2, 3)) # с помощью randint() можно сгенерировать целые числа. Передавая один аргумент 4(не включая) это создает только верхнюю границу диапазона.
```

```python
np.random.randint(6, 12, size=(3, 3)) # первый аргумент 6 будет нижний границей диапазона, а 12 верхний(не включая).
```

```python
arr = np.arange(6)
print(arr)
print(np.random.shuffle(arr)) # функция шафл shuffle() перемешивания массива.
arr
```

```python
playlist = ["The Beatles","Pick Floid","ACDC","Deep Purple","Metalica","RHCP","Slipknot"]

shuffled = np.random.permutation(playlist) # функция permutation() создает новый массив уже перемешанный композиций.
print(shuffled)
print(playlist)
```

```python
np.random.permutation(10) # передаем число которое будет верхней границей для функции. Если хотите перемешивать числа от и до верхней границы.
```

```python
np.random.choice(5, size=(2, 3)) # функция choice() нужна для работы с разными выборками, она позволяет выбрать из массива данных, какое-то подмножество случайным образом.
```

```python
workers = ['Ivan','Nikita','Maria','John','Kate','Rassul','Guram','Patima']

choice = np.random.choice(workers, size=2, replace=False) # replace=False - это параметр, для формирования выборки без повторений.
print(choice)
```

```python
workers = ['Ivan','Nikita','Maria','John','Kate','Rassul','Guram','Patima']

choice = np.random.choice(workers, size=10, replace=False) # Задать выборке размерность и без повторений возможны ошибки, если выборка не достаточно велика.
# невозможно получить элементов без повторений больше, чем их доступно изначально.
print(choice)
```

```python
choice = np.random.choice([1, 2, 3, 4, 5, 6], size=10) # тут получает массивы без всяких проблем.
print(choice)
```

```python
np.random.seed(23) # начальное число для генерации рандомных чисел функция seed().
np.random.randint(10, size=(3, 4))
```

```python
np.random.seed(100)
print(np.random.randint(10, size=(3)))
print(np.random.randint(10, size=(3)))
print(np.random.randint(10, size=(3)))
```

```python
# seed() должен быть >= 0 и при этом не превосходить 2**32-1. Это правило формирования числа для функции seed().
# seed() число что передают в функцию, должно быть формата unsignedint32() т.е. для числа должно быть выделенно 32 бита памяти и при этом данное число не должно иметь знака.
print(2**32-1)
```
