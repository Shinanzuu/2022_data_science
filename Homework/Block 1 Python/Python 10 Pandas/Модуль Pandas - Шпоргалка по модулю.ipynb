```python
# узнаем размер таблицы

melb_data.shape

# Видим, 13580 строк и 23 столбца(параметра).
```

```python
# Получаем детальную информацию о таблице

melb_data.info()
```

```python
# Используем метод Describe() основные статистические характеристики, таким образом выводим только числовые значения

melb_data.describe()
```

```python
# Используя в методе describe(include=['object']) выводим: 

# кол-во непустых строк (count)
# кол-во уникальных значений (unique)
# кол-во частых значений - моду(mode) (top)
# кол-во частоты использования данных значений (freq) для каждого столба исходной таблицы

melb_data.describe(include=['object'])
```

```python
# Частота каждого уникального значения находим с помощью value_counts()

melb_data['Regionname'].value_counts()
```

```python
# С помощью параметра (normalize=True) результат значения будет представлен в виде доли относительной частоты, долях, процентном соотношении где 1 == 100%.

melb_data['Regionname'].value_counts(normalize=True)
```
